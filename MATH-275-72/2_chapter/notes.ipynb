### Definition 4

Let $f: A \rightarrow B$ and let S be a subset of A. The image of S under the function f is a subset
of B that consists of the images of the elements of S. We denote the image of S by $f(S)$, so

$f(S) = \{t | \exists s \in S (t = f(s))\}$

We also use the shorthand $\{f(s) | s \in S\}$ to denote this set

#### Increase and Decreasing

1. Increasing: &emsp;$\forall x \forall y (x<y \rightarrow f(x) \geq f(y))$ )
1. Strictly increasing: &emsp;$\forall x \forall y (x<y \rightarrow f(x) > y(x))$
1. Descreasing: &emsp;$\forall x \forall y (x<y \rightarrow f(x) \leq f(y))$
1. Stricly decreasing: &emsp;$\forall x \forall y (x<y \rightarrow f(x) < f(y))$

A function that is strictly increasing or strictly decreasing must be one to one.

### Definition 7

A function f from A to B is called onto, or a surjection. If and only if for every element $b \in B$ there is an element
$a \in A$ with $f(a) = b$. A function is called surjective if it is onto.

$a \in A \iff b \in B$

A function is onto if $\forall y \exists x (f(x) = y)$, where the domain for x is the domain of the function and the domain of
y is the codomain of the function.

### Definition 8

The function f is a one to one correspondence, or a bijection, if it is both one to one and onto.
We also say that such a function is bijective.


Identity Function = $\iota_a : A \rightarrow A$, where $\iota_a(x)=x$ for all $x \in A$. 

The Identity function is one-to-one and onto, so it is a bijection. 

## Reference

Suppose that $f: A \rightarrow B$

*To show that f is injective* $(y \in A) \rightarrow \{ x | f(x) = f(y) \rightarrow x=y \}$

*To show that f is not injective* Find particular elements x, $y \in A$ such that $x \neq y$ and $f(x) = f(y)$

*To show that f is surjective* Consider an arbitrary element $y \in B$ and find an element $x \in A$ such that
$f(x) = y$

*To show that f is not surjective* $\forall x \in A ( y \in B | f(x) \neq y)$
